```python
df = pd.read_csv("Data/Interpolated_soil_combined/combined_zones.csv")

# Define the function that will assign the soil test
def assign_soil_test(field_id):
    if field_id in ['16A', '6-12', 'SW16']:
        return 'Mehlich'
    elif field_id in ['Y10', 'Y8']:
        return 'Haney'
    else:
        return None  # or some default value

# Create the new column
df['Soil_Test'] = df['FieldID'].apply(assign_soil_test)

rainfall = {2018: 350, 
            2019: 505,
            2020: 481,
            2022: 279}

df['Rainfall'] = df['Year'].map(rainfall)

df = df[['Longitude', 'Latitude',
       'Yield', 'Moisture', 'PlotID', 'FieldID', 'TOC_0_2', 'IC_0_2', 'TN_0_2', 
                           'DI_Al_0_2', 'DI_Ca_0_2', 'DI_Fe_0_2', 'DI_K_0_2', 'DI_Mg_0_2', 'DI_P_0_2', 
                           'H3A_Al_0_2', 'H3A_Ca_0_2', 'H3A_Fe_0_2', 'H3A_K_0_2', 'H3A_Mg_0_2', 'H3A_P_0_2',
                           'M3_Al_0_2', 'M3_Ca_0_2', 'M3_Fe_0_2', 'M3_K_0_2', 'M3_Mg_0_2', 'M3_P_0_2',
                           'Ols_Al_0_2', 'Ols_Ca_0_2', 'Ols_Fe_0_2', 'Ols_K_0_2', 'Ols_Mg_0_2', 'Ols_P_0_2',
                           'TOC_2_6', 'IC_2_6', 'TN_2_6',
                           'DI_Al_2_6', 'DI_Ca_2_6', 'DI_Fe_2_6', 'DI_K_2_6', 'DI_Mg_2_6', 'DI_P_2_6', 
                           'H3A_Al_2_6', 'H3A_Ca_2_6', 'H3A_Fe_2_6', 'H3A_K_2_6', 'H3A_Mg_2_6', 'H3A_P_2_6', 
                           'M3_Al_2_6', 'M3_Ca_2_6', 'M3_Fe_2_6', 'M3_K_2_6', 'M3_Mg_2_6', 'M3_P_2_6',
                           'Ols_Al_2_6', 'Ols_Ca_2_6', 'Ols_Fe_2_6', 'Ols_K_2_6', 'Ols_Mg_2_6', 'Ols_P_2_6',
                           'Rainfall', 'PhosphorusTreatment',
        'Zone_1', 'Zone_2', 'Zone_3', 'Zone_4', 'Year']]

df['Yield'] = df['Yield']* 25.4016 # Converting to kg/ha
df = df[df['PhosphorusTreatment'] == 0]

#'TC_0_2', 'TC_2_6', 'elevation', 'slope','aspect', 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define a function to create the plots
def plot_median_soil_test_vs_yield_by_year_field_zone(df, soil_test_parameter):
    # Group by Year, FieldID, and Zone, and calculate median
    grouped = df.groupby(['Year', 'FieldID', 'Zone_1', 'Zone_2', 'Zone_3', 'Zone_4']).median().reset_index()

    # Plot
    plt.figure(figsize=(12, 6))
    sns.scatterplot(data=grouped, x=soil_test_parameter, y='Yield', hue='FieldID', style='Year', size='Year', alpha=0.7)
    plt.title(f'Median {soil_test_parameter} vs Yield by Year, Field, and Zone')
    plt.xlabel(f'Median {soil_test_parameter}')
    plt.ylabel('Yield')
    plt.grid(True)
    plt.legend(title='Field and Year', bbox_to_anchor=(1.05, 1), loc=2)
    plt.show()
```

```python
# Create the plots for each soil test parameter
plot_median_soil_test_vs_yield_by_year_field_zone(df, 'Ols_P_0_2')
plot_median_soil_test_vs_yield_by_year_field_zone(df, 'H3A_P_0_2')
plot_median_soil_test_vs_yield_by_year_field_zone(df, 'M3_P_0_2')
```

# Plotting relative yield

```python
# Define a function to create the plots with relative yield by year and field
def plot_median_soil_test_vs_relative_yield_by_year_field_zone(df, soil_test_parameter):
    # Calculate the maximum yield for each combination of year and field
    max_yield_by_year_field = df.groupby(['Year', 'FieldID'])['Yield'].max()

    # Calculate the relative yield as a percentage of the maximum yield of that year and field
    df['Relative_Yield'] = df.apply(lambda row: (row['Yield'] / max_yield_by_year_field[row['Year'], row['FieldID']]) * 100, axis=1)

    # Group by Year, FieldID, and Zone, and calculate median
    grouped = df.groupby(['Year', 'FieldID', 'Zone_1', 'Zone_2', 'Zone_3', 'Zone_4']).median().reset_index()

    # Plot
    plt.figure(figsize=(12, 6))
    sns.scatterplot(data=grouped, x=soil_test_parameter, y='Relative_Yield', hue='FieldID', style='Year', size='Year', alpha=0.7)
    plt.title(f'Median {soil_test_parameter} vs Relative Yield by Year, Field, and Zone')
    plt.xlabel(f'Median {soil_test_parameter}')
    plt.ylabel('Relative Yield (%)')
    plt.grid(True)
    plt.legend(title='Field and Year', bbox_to_anchor=(1.05, 1), loc=2)
    plt.show()
```

```python
# Create the plots for each soil test parameter
plot_median_soil_test_vs_relative_yield_by_year_field_zone(df, 'Ols_P_0_2')
plot_median_soil_test_vs_relative_yield_by_year_field_zone(df, 'H3A_P_0_2')
plot_median_soil_test_vs_relative_yield_by_year_field_zone(df, 'M3_P_0_2')
```

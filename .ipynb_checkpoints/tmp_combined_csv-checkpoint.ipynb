```python
import pandas as pd
```

```python
data = pd.read_csv('HPC_code/Data/combined_zones.csv')
data.head()
```

```python
data.columns.values
```

```python
# List of zone columns (adjust as per your actual column names)
zone_columns = ['Zone_1', 'Zone_2', 'Zone_3', 'Zone_4']

# Initialize an empty DataFrame to store results
zone_counts = pd.DataFrame()

# Count occurrences for each zone
for zone in zone_columns:
    # Summing the zone column grouped by 'FieldID'
    zone_count = data.groupby('FieldID')[zone].sum().reset_index(name=f'count_{zone}')
    zone_counts = pd.concat([zone_counts, zone_count.set_index('FieldID')], axis=1)

# Reset index to make 'FieldID' a column again
zone_counts = zone_counts.reset_index()

zone_counts
```
